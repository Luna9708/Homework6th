```python
dta = pd.DataFrame([data]).T
```

```python
dta
```

```python
fig = plt.figure(figsize=(12,8))
ax1=fig.add_subplot(211)
# 绘制ACF图
fig = sm.graphics.tsa.plot_acf(dta,lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
# 绘制PACF图
fig = sm.graphics.tsa.plot_pacf(dta,lags=40,ax=ax2)
plt.show()
```

```python
arma_mod70 = sm.tsa.ARMA(dta,(7,0)).fit()
#print(arma_mod70.aic)
```

```python
predict_dta = arma_mod70.predict(90,94,dynamic=True)
print(predict_dta)
fig, ax = plt.subplots(figsize=(12,8))
ax = dta.ix[0:].plot(ax=ax)

fig = arma_mod70.plot_predict(89,94,dynamic=True, ax=ax, plot_insample=False)
plt.show()
```
